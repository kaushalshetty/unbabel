```
import re
def reverseWords(input_text):      
    inputWords = input_text.split(" ")  
    inputWords=inputWords[-1::-1]
    starting_word = None
    ending_word = None
    caps_word_cnt = 0
    for word in inputWords:
      if len(word)>0:
        if word[0].isupper():
          starting_word = word
          caps_word_cnt += 1
        elif re.match('[a-z][^?!.]*[?.!]', word) is not None:
          ending_word = word
    if ending_word is not None:      
      inputWords.remove(ending_word)      
    if caps_word_cnt == 1:
          inputWords.remove(starting_word)
          output = starting_word + ' ' + ' '.join(inputWords)
    else:
      output = ' '.join(inputWords)      
    if ending_word is not None:
      output = output + ' '+ending_word
    return output
```

```
input_file = open('data/train_scrambled.txt', 'r') 
train_data = input_file.readlines()
```

```
unscrambled_baseline = [reverseWords(item.strip()) for item in train_data]
```
